```python

# Read data from CSV file
data = pd.read_csv("Twiter_Data.csv")

# Select the column with the text data
text_data = data["Data"]

# Tokenize the text
tokens = [word_tokenize(post) for post in text_data]

# Remove stop words
stop_words = set(stopwords.words("english"))
filtered_tokens = [[word for word in post if word.lower() not in stop_words] for post in tokens]

# Perform stemming
ps = PorterStemmer()
stemmed_tokens = [[ps.stem(word) for word in post] for post in filtered_tokens]

# Print the stemmed tokens
print(stemmed_tokens)
```
